```python
# Build compact database

import sqlite3 
import os
import numpy as np
import pandas as pd

my_dir = os.path.realpath('.')
db_file = os.path.join(my_dir, '../../gr_sentiment_analysis/data/books.db')
db_compact = os.path.join(my_dir, '../../gr_sentiment_analysis/data/books_compact.db')
conn = sqlite3.connect(db_file)
conn_compact = sqlite3.connect(db_compact)

```

```python
# Get a sample of reviews. Say, 10%
review_id_qry = 'SELECT review_id FROM review_stats WHERE word_count > 30 AND afinn_mean IS NOT NULL AND bing_mean IS NOT NULL AND mpqa_mean IS NOT NULL AND inq_mean IS NOT NULL'

review_ids = pd.read_sql(review_id_qry, con=conn)
rev_ids = review_ids.sample(frac=.1)['review_id']
rev_ids = ', '.join(map(str, list(rev_ids)))
```

```python
reviews_qry = 'SELECT * FROM reviews WHERE review_id IN ({0})'.format(rev_ids)
reviews = pd.read_sql(reviews_qry, conn)
reviews.to_sql('reviews', con=conn_compact, if_exists='replace', index=False)
```

```python
book_ids_qry = 'SELECT book_id FROM reviews WHERE review_id IN ({0})'.format(rev_ids)
book_ids = pd.read_sql(book_ids_qry, con=conn)['book_id']
book_ids = ', '.join(map(str, list(book_ids)))
```

```python
afinn = pd.read_sql("select * from afinn_lexicon", conn)
afinn.to_sql('afinn_lexicon', con=conn_compact, if_exists='replace', index=False)

bing = pd.read_sql("select * from bing_lexicon", conn)
bing.to_sql('bing_lexicon', con=conn_compact, if_exists='replace', index=False)

inq = pd.read_sql("select * from inquirer_lexicon", conn)
inq.to_sql('inquirer_lexicon', con=conn_compact, if_exists='replace', index=False)

mpqa = pd.read_sql("select * from mpqa_lexicon", conn)
mpqa.to_sql('mpqa_lexicon', con=conn_compact, if_exists='replace', index=False)
```

```python
book_info_qry = 'SELECT * FROM book_info WHERE id IN ({0})'.format(book_ids)
book_info = pd.read_sql(book_info_qry, conn)
book_info.to_sql('book_info', con=conn_compact, if_exists='replace', index=False)
```

```python
book_info_clean_qry = 'SELECT * FROM book_info_clean WHERE id IN ({0})'.format(book_ids)
book_info_clean = pd.read_sql(book_info_clean_qry, conn)
book_info_clean.to_sql('book_info_clean', con=conn_compact, if_exists='replace', index=False)
```

```python
review_words_qry = 'SELECT * FROM review_words WHERE review_id IN ({0})'.format(rev_ids)
review_words = pd.read_sql(review_words_qry, conn)
review_words.to_sql('review_words', con=conn_compact, if_exists='replace', index=False)
```

```python
review_stats_qry = 'SELECT * FROM review_stats WHERE review_id IN ({0})'.format(rev_ids)
review_stats = pd.read_sql(review_stats_qry, conn)
review_stats.to_sql('review_stats', con=conn_compact, if_exists='replace', index=False)
```

```python
review_features_qry = 'SELECT * FROM review_features WHERE review_id IN ({0})'.format(rev_ids)
review_features = pd.read_sql(review_features_qry, conn)
review_features.to_sql('review_features', con=conn_compact, if_exists='replace', index=False)
```

```python
conn_compact.close()
```
